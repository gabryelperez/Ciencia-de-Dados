```python
# Importa a biblioteca Pandas
import pandas as pd
```

```python
# Importa o arquivo em excel, ignorando as 3 primeiras linhas e as 3 últimas linhas
df = pd.read_excel("201710_EmissaoPassagens_SCDP.xlsx",  skiprows=3, skipfooter=3)
```

```python
# Lista os cinco primeiros registros
df.head()
```

```python
# Lista os cinco últimos registros
df.tail()
```

```python
# Renomeia as colunas removendo os espaços em branco a esquerda
df.columns = (['Código do órgão superior', 'Nome do órgão superior', 'Código do órgão solicitante da viagem', 
               'Nome do órgão solicitante da viagem', 'N. PCDP', 'N. Reserva/Localizador', 
               'Data Emissão Bilhete', 'Data Embarque', 'Valor Tarifa Comercial', 'Percentual Desconto Aplicado', 
               'Valor Tarifa Governo', 'Valor Tarifa Embarque', 'Valor Bilhete', 'Companhia Aérea', 
               'Classe Tarifária Bilhete', 'Regra Tarifária', 'No Show', 'Remarcado', 'Cancelado', 'Valor Multas', 
               'Valor Reembolso', 'Diferença de Tarifa', 'Situação Final Bilhete'])
```

```python
# Verifica os tipos de dados das variáveis
df.dtypes
```

```python
# Converte a coluna de Data Emissão Bilhete e Data Embarque para Datetime
df["Data Emissão Bilhete"] = pd.to_datetime(df["Data Emissão Bilhete"])
df["Data Embarque"] = pd.to_datetime(df["Data Embarque"])
```

```python
# Verifica se as conversões foram realizadas
df.dtypes
```

```python
# Estatísticas Descritivas do Data Frame
df.describe()
```

```python
# Quantidade Emissão por Ano
print("Quantidade de Emissão por Ano")
print(df["Data Emissão Bilhete"].dt.year.value_counts())
```

```python
# Quantidade de Emissão por Mês
print("Quantidade de Emissão por Mês")
print(df["Data Emissão Bilhete"].dt.month.value_counts())
```

```python
# Quantidade de Embarque por Ano
print("Quantidade de Embarque por Ano")
print(df["Data Embarque"].dt.year.value_counts())
```

```python
# Quantidade de Embarque por Mês
print("Quantidade de Embarque por Mês")
print(df["Data Embarque"].dt.month.value_counts())
```

```python
# Importando a biblioteca Matplotlib
import matplotlib.pyplot as plt
```

```python
# Gráfico da Quantidade de Passagens por Situação Final do Bilhete
df['Situação Final Bilhete'].value_counts().plot.pie()
plt.title('Quantidade de Passagens por Situação Final do Bilhete')
plt.legend()
plt.show()
```

```python
# Gráfico da Quantidade de Passagens Canceladas
df['Cancelado'].value_counts().plot.bar()
plt.title('Quantidade de Passagens Canceladas')
plt.show()
```

```python
# Gráfico da Quantidade de Passagens Remarcadas
df['Remarcado'].value_counts().plot.bar()
plt.title('Quantidade de Passagens Remarcadas')
plt.show()
```

```python
# Gráfico da Quantidade de Passagens No Show
df['No Show'].value_counts().plot.bar()
plt.title('Quantidade de Passagens No Show')
plt.show()
```

```python
# Gráfico da Quantidade de Passagens por Companhia Aérea
df['Companhia Aérea'].value_counts().plot.barh()
plt.grid(True, linestyle='--')
plt.title('Quantidade de Passagens por Companhia Aérea')
plt.xlabel('Companhia Aérea')
plt.ylabel('Quantidade de Passagens')
plt.show()
```
